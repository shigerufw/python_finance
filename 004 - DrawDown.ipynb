### 3.1 Wealth index

```python
wealth_m = 100 #pct
wealth_index = wealth_m*df_cumprod
wealth_index.plot()
```

### 3.2 Previous peaks

```python
previous_peaks = wealth_index.cummax()
previous_peaks.plot()
```

### 3.3 - Drawdown

```python
drawdown = (wealth_index- previous_peaks)/previous_peaks
drawdown.plot()
```

# 4- Creating a function to help 

```python
def drawdown(return_series):
    """
    
    input: time series of asset pct returns
    output: (dataframe of returns , previous peaks, percent drawdowns)
    """
    wealth_index = 100*(1+return_series).cumprod()
    previous_peaks = wealth_index.cummax()
    drawdowns = (wealth_index-previous_peaks)/previous_peaks
    return pd.DataFrame({
        "Wealth":wealth_index,
        "Peaks":previous_peaks,
        "Drawdowns": drawdowns
    })
    
```

#### Plotting 

```python
drawdown(df_pct_change["Nasdaq"])[['Drawdowns']].min()*100
```

```python
plt.figure(figsize = (12,8))
#Wealth
plt.plot(drawdown(df_pct_change["Nasdaq"])[['Wealth']], label = ['Weath'])
#Peaks
plt.plot(drawdown(df_pct_change["Nasdaq"])[['Peaks']], label = ['Peaks'])
#min Drawndow

plt.vlines(x=drawdown(df_pct_change["Nasdaq"])[['Drawdowns']].idxmin(),
          ymin= drawdown(df_pct_change["Nasdaq"])[['Wealth']].min(),
          ymax= drawdown(df_pct_change["Nasdaq"])[['Peaks']].max(),        
          color = 'r',)

#annotate
plt.annotate("max", [drawdown(df_pct_change["Nasdaq"])[['Drawdowns']].idxmin(),drawdown(df_pct_change["Nasdaq"])[['Peaks']].max()])
plt.annotate("min", [drawdown(df_pct_change["Nasdaq"])[['Drawdowns']].idxmin(),drawdown(df_pct_change["Nasdaq"])[['Wealth']].min()])
plt.legend()
```
